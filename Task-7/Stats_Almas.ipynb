### 2) MEDIAN
It is also one of the measure of central tendencyin given set of values.
<!--spaces-->
Median is the value separating the higher half from the lower half of a data distribution.
<!-- spaces -->
Eg: if [2,4,77,1,9] is given 
<!--spaces-->
Median is 4
<!--spaces-->
[1,2,4,9,77]  
<!-- spaces -->
Steps to find median:
<!-- spaces -->
 Sort the elements in given set
<!-- spaces -->
 If the number of elements n is odd,then median is $$ (\frac{n}{2}+1)^{th}\, value\, in\, sorted\, set.$$
<!-- spaces -->
 If the number of elements n is even, then median is $$  ((\frac{n}{2}+1)+(\frac{n}{2}))^{th}\, value\, in\, sorted\, set.$$
<!--spaces-->

<!--spaces-->
Median is defined as the center value of ordered list of values
<!--spaces-->
The median is the most informative measure of central tendency for skewed distributions or distributions with outliers.

##### Finding median without using python libraries

```python
#Considering Mthly_HH_Income
mthly_ex = data['Mthly_HH_Income']
n = len(mthly_ex)
mthly_ex.sort_values()
if n % 2 == 0:
    median1 = mthly_ex[n//2]
    median2 = mthly_ex[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = mthly_ex[n//2]
print("Median of 'Mthly_HH_Income' column values :",median)
```

##### Finding median using numpy library

```python
#Using numpy library
mthly_ex = data['Mthly_HH_Income']
med = np.median(mthly_ex)
print("Median of 'Mthly_HH_Income' column values using numpy library median function :",med)
```

### 4)VARIANCE
Variance is a measure of dispersion that takes into account the spread of all data points in a data set.
<!--space-->
The variance for a sample is calculated by
<!--space-->
 Finding the mean.
<!--space-->
 Subtracting the mean from each number in the data set.
<!--space-->
 Squaring the result
<!--space-->
 The results are squared to make the negatives positive.
<!--space-->
$$ S^{2} = \frac{\Sigma(x_{i}-\bar{x})^{2}}{n-1}  $$
<!--spaces-->

<!--spaces-->
The advantage of variance is that it treats all deviations from the mean as the same regardless of their direction.
<!--spaces-->
Variance helps to find the distribution of data in a population from a mean.


<!--space-->
##### Finding variance without using python libraries.

```python
# Considering Emi_or_Rent_Amt column
emi_r = data['Emi_or_Rent_Amt']
avg = sum(emi_r) / len(emi_r)
var = sum((x-avg)**2 for x in emi_r) / len(emi_r)
print("Variance of 'Emi_or_Rent_Amt' column values is : ",var)
```

##### Finding variance using numpy library

```python
#using numpy library
var = np.var(emi_r)
print("Variance of 'Emi_or_Rent_Amt' column values is : ",var)
```

### 5)STANDARD DEVIATION
Standard deviation is a statistic that measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance.
<!--space-->
Standard deviation is a number that describes how spread out the values are.
<!--space-->
A low standard deviation means that most of the numbers are close to the mean (average) value.
<!--space-->
A high standard deviation means that the values are spread out over a wider range.
<!--space-->
Standard deviation is calculated as the square root of the variance. Alternatively, it is calculated by finding the mean of a data set, finding the difference of each data point to the mean, squaring the differences, adding them together, dividing by the number of points in the data set less 1, and finding the square root.
<!--space-->
$$ S = \sqrt{\frac{\Sigma(x_{i}-\bar{x})^{2}}{n-1}}  \\ \\$$
$$ \sigma = \sqrt{var} $$
<!--space-->

<!--space-->
Standard deviation gives more clarity about the deviation of data from a mean.
<!---->
standard deviation also helps to know the distribution of data in population.
<!--space-->
Standard deviation is important because it tells us how spread out the values are in a given dataset.


<!--space-->
#####  Finding standard deviation without using python library

```python
#Considering Annual_HH_Income 

ann_hh = data['Annual_HH_Income']
mean = sum(ann_hh) / len(ann_hh)   # mean
var  = sum(pow(x-mean,2) for x in ann_hh) / len(ann_hh)  # variance
std  = math.sqrt(var) # standard deviation
print("Standard deviation of 'Annual_HH_Income' column values is :",std)
```

##### Finding standard deviation using numpy library

```python
#using library

std = np.std(ann_hh)
print("Standard deviation of 'Annual_HH_Income' column values is :",std)
```

### 6)CORRELATION
Correlation is a statistical measure that expresses the extent to which two variables are linearly related.
<!--space-->
Correlation measures the relationship between two variables.
<!--space-->
Correlation Coefficient
<!--space-->
The correlation coefficient measures the relationship between two variables.
<!--space-->
The correlation coefficient can never be less than -1 or higher than 1.
<!--space-->
 1 = there is a perfect linear relationship between the variables.
<!--space-->
 0 = there is no linear relationship between the variables.
<!--space-->
 -1 = there is a perfect negative linear relationship between the variables.
<!--space-->
Let $(_{i}, _{i})$ = ordered pairs that represent points on a scatter plot.
<!--space-->
$\bar{x}$= mean of the  values
<!--space-->
$\bar{}$ = mean of the  values
<!--space-->
$S_{x}$ = standard deviation of  values
<!--space-->
$S_{y}$ = standard deviation of  values
<!--space-->
Correlation Co-efficeint is 
<!--space-->
$ r = \frac{1}{n-1}\sum_{i=1}^{n}(\frac{x_{i}-\bar{x}}{S_{x}})(\frac{y_{i}-\bar{y}}{S_{y}}) $
<!--space-->
Pearson's Correlation Coefficient
<!--space-->
$ r = \frac{\sum_{i=1}^{n}(x_{i}-\bar{x})(y_{i}-\bar{y})}{\sqrt{\sum_{i=1}^{n}}(x_{i}-\bar{x})^{2}\sqrt{\sum_{i=1}^{n}}(y_{i}-\bar{y})^{2}} $
<!--space-->
$ r = \frac{n(\Sigma xy) - (\Sigma x)(\Sigma y)}{\sqrt{[n\Sigma x^{2} - (\Sigma x)^2][n\Sigma y^2 - (\Sigma y)^2]}} $
<!--space-->
There are two main types of correlation coefficients: Pearson's product moment correlation coefficient and Spearman's rank correlation coefficient
<!--space-->

<!--space-->
In statistical terms, correlation is a method of assessing a possible two-way linear association between two continuous variables.1 Correlation is measured by a statistic called the correlation coefficient, which represents the strength of the putative linear association between the variables in question.
<!--space-->


##### Finding correlation coefficient without using libraries
'No_of_Earning_Members' and  'Mthly_HH_Income'

```python
# function that returns correlation coefficient.
def correlationCoefficient(X, Y, n) :
    sum_X = 0
    sum_Y = 0
    sum_XY = 0
    squareSum_X = 0
    squareSum_Y = 0
      
      
    i = 0
    while i < n :
        # sum of elements of array X.
        sum_X = sum_X + X[i]
          
        # sum of elements of array Y.
        sum_Y = sum_Y + Y[i]
          
        # sum of X[i] * Y[i].
        sum_XY = sum_XY + X[i] * Y[i]
          
        # sum of square of array elements.
        squareSum_X = squareSum_X + X[i] * X[i]
        squareSum_Y = squareSum_Y + Y[i] * Y[i]
          
        i = i + 1
       
    # use formula for calculating correlation 
    # coefficient.
    corr = (float)(n * sum_XY - sum_X * sum_Y)/(float)(math.sqrt((n * squareSum_X - sum_X * sum_X)* (n * squareSum_Y - sum_Y * sum_Y)))
    return corr
      
# Driver function
X = data['No_of_Earning_Members']
Y = data['Mthly_HH_Income']
       
# Find the size of array.
n = len(X)
  
# Function call to correlationCoefficient.
print ('{0:.6f}'.format(correlationCoefficient(X, Y, n)))
  
```

In above example correlation coefficient obtained is 0.347883 
<!--space-->
Correlation coefficients whose magnitude are between 0.3 and 0.5 indicate variables which have a low correlation.

### 7)NORMAL DISTRIBUTION
History of Normal Distribution:
<!--space-->
Carl Friedrich Gauss (17771855) Referred to as the "Prince of Mathematicians"
<!--space-->
Credited with the use of the probability distribution now known as the normal or Gaussian distribution(bell curve)
<!--space-->
The normal distribution:
<!--space-->
$$ f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{\frac{-1}{2}(\frac{x-\mu}{\sigma})^{2}} $$
$ for -\infty < x < \infty\,
e = 2.7183 \, \pi = 3.1416$
$\mu \,and \,\sigma \,are \,the \,population \,mean\, and\, standard \,deviation $
<!--space-->
 Two parameters, mean and standard deviation, completely determine the Normal
distribution.
<!--space-->
 The shape and location of the normal curve changes as the mean and standard
deviation change.
<!--space-->
Normal distribution is defined by its mean and standard dev.
<!--space-->
$$ E(X) = \mu = \int_{-\infty}^{+\infty} x\frac{1}{\sigma\sqrt{2\pi}}e^{\frac{-1}{2}(\frac{x-\mu}{\sigma})^{2}} \,dx $$
<!--space-->
$$ Var(X) = \sigma^{2} = \int_{-\infty}^{+\infty} x^{2}\frac{1}{\sigma\sqrt{2\pi}}e^{\frac{-1}{2}(\frac{x-\mu}{\sigma})^{2}} \,dx $$
The properties of normal distribution
<!--space-->
 It is a bell-shaped curve.
<!--space-->
 It is symmetrical about the mean, .
<!--space-->
 (The mean, the mode and the median all have the same value).
<!--space-->
 The total area under the curve is 1 (or 100%).
<!--space-->
 50% of the area is to the left of the mean, and 50% to the right.
<!--space-->
The Standard normal distribution (Z)
<!--space-->
 All normal distributions can be converted into the standard normal 
curve by subtracting the mean and dividing by the standard deviation: 
<!--space-->
$ Z = \frac{X - \mu}{\sigma} $
<!--space-->

### 9) POSITIVELY SKEWED AND NEGATIVELY SKEWED NORMAL DISTRIBUTION
Positive skew refers to a longer or fatter tail on the right.
Negative skew refers to a longer or fatter tail on the left side of the distribution.
<!--space-->
 A negative value for skewness indicates that the tail is on the left side of the distribution, which extends towards more negative values.
<!--space-->
 A positive value for skewness indicates that the tail is on the right side of the distribution, which extends towards more positive values.
<!--space-->
Examples of Right-Skewed Distributions/Positive Skewed Distributions:
<!--space-->
Right skewed distributions are the more common form. These distributions tend to occur when there is a lower limit, and most values are relatively close to the lower bound. Values cant be less than this bound but can fall far from the peak on the high end, causing them to skew positively.
<!--space-->
For example, right skewed distributions can occur in the following cases:
<!--space-->
 Time to failure cannot be less than zero, but there is no upper bound.

 Wait and response times cannot be less than zero, but there are no upper limits.

 Sales data cannot be less than zero but can have unusually large values.

 Humans have a minimum viable weight but can have large extreme values.

 Income cannot be less than zero, but there are some extremely high incomes.

Classical example is income and wealth.
<!--space-->
Examples of Left-Skewed Distributions/Negative Skewed Distributions:
<!--space-->
Left skewed distributions occur less frequently than their right-handed counterparts, but they exist. Frequently, they occur when there is an upper limit that values cannot exceed, and most scores are near that limit. Values cant exceed the cap, but they can extend relatively far from the peak on the lower side, causing a negative skew.
<!--space-->
For example, left skewed distributions can occur in the following cases:
<!--space-->
 Purity cannot exceed 100%, but there is room on the low side for extreme values.

 Maximum test scores cannot exceed 100%.

 Ages of death tend to occur around 70-80. Its possible to live a little longer, but extreme values are more likely to appear on the lower end.

<!--space-->

<!--space-->
Right skewed: The mean is greater than the median. The mean overestimates the most common values in a positively skewed distribution.
<!--space-->
Left skewed: The mean is less than the median. The mean underestimates the most common values in a negatively skewed distribution.
<!--space-->
Because the mean over or underestimates the most frequently occurring values in asymmetric distributions, analysts often use the median in these cases. The median is a more robust statistic in the presence of extreme values.

### 11)Explain QQ Plot and show the implementation of the same. 

Q-Q Plot
<!--space-->
 The points Q1, . . . , Qn are called quantiles (divides distribution into equal sized areas) of the distribution.

 These are the points in the data below which a certain proportion of the data falls.

 The probability plot is sometimes called a quantilequantile plot, or QQ plot.

 We can use this Q-Q plot to check the assumption of Normality of the data.

 Determines whether if two set of quantiles come from the populations of same distribution. If, yes roughly forms a straight
line.
<!--space-->
The Quantile-Quantile plot is used for determine whether two samples are from the same population,whether they have the same tail,they have the same distribution shape,they have common location behavior.
<!--space-->
How to create a QQ Plot for a given dataset
<!--space-->
 Collect the data for plotting the quantile-quantile plot.

 Sort the data in ascending or descending order.

 Draw a normal distribution curve.

 Find the z-value (cut-off point) for each segment.

 Plot the dataset values against the normalizing cut-off points.

### 12)Explain Box Cox and show the implementation of the same.
Box Cox Transformation
<!--space-->
A Box Cox transformation is a transformation of non-normal dependent variables into a normal shape. Normality is an important assumption for many statistical techniques; if your data isnt normal, applying a Box-Cox means that you are able to run a broader number of tests.
<!--space-->
The Box Cox transformation is named after statisticians George Box and Sir David Roxbee Cox who collaborated on a 1964 paper and developed the technique.
<!--space-->
At the core of the Box Cox transformation is an exponent, lambda (), which varies from -5 to 5. All values of  are considered and the optimal value for your data is selected; The optimal value is the one which results in the best approximation of a normal distribution curve. The transformation of Y has the form:
<!--space-->
$$ y(\lambda) = \Bigg\{  \frac{y^{\lambda}-1}{\lambda}, if \lambda \neq 0; \  \log y, if \lambda = 0.  $$

This test only works for positive data. However, Box and Cox did propose a second formula that can be used for negative y-values:
<!--space-->
$$ y(\lambda) = \Bigg\{ \frac{(y+\lambda_{2})^{\lambda_{1}} - 1}{\lambda_{1}} \ , if \lambda_{1} \neq 0; \   \log(y + \lambda_{2}) ,\   if \lambda_{1} = 0.   $$

SciPys stats package provides a function called boxcox for performing box-cox power transformation that takes in original non-normal data as input and returns fitted data along with the lambda value that was used to fit the non-normal distribution to normal distribution.




```python
ah_in = data['Annual_HH_Income']
 
# transform training data & save lambda value
fitted_data, fitted_lambda = stats.boxcox(ah_in)
 
# creating axes to draw plots
fig, ax = plt.subplots(1, 2)
 
# plotting the original data(non-normal) and fitted data (normal)
sns.distplot(ah_in, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Non-Normal", color ="green", ax = ax[0])
 
sns.distplot(fitted_data, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Normal", color ="green", ax = ax[1])
 
# adding legends to the subplots
plt.legend(loc = "upper right")
 
# rescaling the subplots
fig.set_figheight(5)
fig.set_figwidth(10)
 
print(f"Lambda value used for Transformation: {fitted_lambda}")
```

### Pareto Distribution
Pareto Distribution is a power-law property.
If X is a random variable with a Pareto (Type I) distribution, then the probability that X is greater than some number x, i.e. the survival function (also called tail function), is given by
 $ \bar{F}(x) = Pr(X > x) = ={\begin{cases}\left({\frac {x_{\mathrm {m} }}{x}}\right)^{\alpha }&x\geq x_{\mathrm {m} },\\1&x<x_{\mathrm {m} },\end{cases}} \\ $
where $x_{m} $ is the (necessarily positive) minimum possible value of X, and  is a positive parameter. The Pareto Type I distribution is characterized by a scale parameter $ x_{m} $ and a shape parameter , which is known as the tail index. When this distribution is used to model the distribution of wealth, then the parameter  is called the Pareto index.

### Logn Distribution
A log-normal distribution is a statistical distribution of logarithmic values from a related normal distribution. A log-normal distribution can be translated to a normal distribution and vice versa using associated logarithmic calculations.

### Bernoulli Distribution
A Bernoulli distribution is a discrete probability distribution for a Bernoulli trial  a random experiment that has only two outcomes.
 usually called a Success or a Failure


### Binomial Distribution
 The Binomial distribution is closely related to the Bernoulli distribution

 A Binomial Distribution can be simply defined as a sequence of Bernoulli trials each independent of one another.

 In general, if there are n Bernoulli trials, then the sum of those trials is binomially distributed with parameters n and p
 
             X ~ Bin(n,p)

### Poisson Distribution
Poisson distribution is as an approximation to the binomial distribution when n is large and p is small
 If the numbers of events in disjoint intervals of time or space are independent, and if events
cannot occur simultaneously, then X follows a Poisson distribution.

 A process that produces such events is called a Poisson process. Since the mean number of
events that occur in t units of time or space is equal to t
                X ~ Poisson(t)
